## KDE Plot 

```python
sns.kdeplot(df['Sales'],color='red',label='Sales',shade=True)
sns.kdeplot(df['Profit'],color='Blue',label='Profit',shade=True)
plt.xlim([-100,1000])
plt.show()
```

## Pairplot Analysis

```python
#Based on category
sns.pairplot(df,hue='Category')
```

```python
#Based on Region
sns.pairplot(df,hue='Region')
```

## Correlation Matrix 

```python
#to check correlationship between the variables
df1=df.corr()
plt.figure(figsize=(10,6))
sns.heatmap(df1,annot=True)
plt.show()
```

## Count Plot 

```python
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
sns.countplot(df['Category'],ax=axs[0][0])
sns.countplot(df['Segment'],ax=axs[0][1])
sns.countplot(df['Region'],ax=axs[1][0])
sns.countplot(df['Ship Mode'],ax=axs[1][1])
axs[0][0].set_title('Category')
axs[0][1].set_title('Segment')
axs[1][0].set_title('Region')
axs[1][1].set_title('Ship Mode')
```

```python
plt.figure(figsize=(20,10))
sns.countplot(df['Sub-Category'])
plt.title('Sub-Category',fontsize=10)
```

```python
plt.figure(figsize=(20,20))
sns.countplot(df['State'])
plt.xticks(rotation=90)
plt.title('State',fontsize=20)
```

```python
plt.figure(figsize=(20,10))
sns.countplot(df['Quantity'])
plt.title('Quantity',fontsize=20)
```

```python
plt.figure(figsize=(20,10))
sns.countplot(df['Discount'])
plt.title('Discount',fontsize=10)
```

## Group wise Analysis 

```python
df['Segment'].value_counts()
```

```python
df_segment=df.groupby(['Segment'])[['Sales','Discount','Profit']].mean()
df_segment
```

```python
df_state=df.groupby(['State'])[['Sales','Discount','Profit']].mean()
df_state.head(10)
```

```python
df_state1=df_state.sort_values('Profit')
df_state1[['Profit']].plot(kind='bar',figsize=(15,5))
plt.title('State Wise Profit Analysis',fontsize=20)
plt.xlabel('States')
plt.ylabel('Profit per State')
plt.show()
```
